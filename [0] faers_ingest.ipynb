### Fix a record with no manufacturers in Demographic table to prevent "None" manufacturer record

```python
%%bigquery
select distinct mfr_sndr from faers_staging.Demographic
order by mfr_sndr
# View unique drug manufacturers - "None" manufacturer record exists
```

```python
%%bigquery
UPDATE faers_staging.Demographic
SET mfr_sndr = 'FDA-CTU'
WHERE mfr_sndr is null
```

```python
%%bigquery
select distinct mfr_sndr from faers_staging.Demographic
order by mfr_sndr
# None manufacturer no longer exists
```

### Fix age record errors in Demographic table

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'HR'
# Age is 44 hours but person is adult -> change HR to YR 
```

```python
%%bigquery
UPDATE faers_staging.Demographic
SET age_cod = 'YR'
WHERE age_cod = 'HR' and age_grp='A'
```

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'HR' and age_grp='A'
# Record fixed
```

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'WK' and age_grp is not null
order by age_grp 
# Age is 42 weeks but person is adult -> change WK to YR 
```

```python
%%bigquery
UPDATE faers_staging.Demographic
SET age_cod = 'YR'
WHERE age_cod = 'WK' and age_grp='A'
```

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'WK' and age_grp='A'
# Record fixed
```

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'DY' and age_grp is not null
order by age_grp
limit 5
# Age is 0 days but person is adult -> change A to N (Neonatal) 
```

```python
%%bigquery
UPDATE faers_staging.Demographic
SET age_grp = 'N'
WHERE age_cod = 'DY' and age_grp='A'
```

```python
%%bigquery
select primaryid, age, age_cod, age_grp
from faers_staging.Demographic
WHERE age_cod = 'DY' and age_grp='A'
# Record fixed
```

### Replace 'None' record of PROD_AI (Product Active_Ingredient) in Drug table with 'UNKNOWN' to avoid null records of active ingredients 

```python
%%bigquery
select count(*) from faers_staging.Drug
where PROD_AI is null
```

```python
%%bigquery
UPDATE faers_staging.Drug
SET PROD_AI = 'UNKNOWN'
WHERE PROD_AI is null
```
