### Transforming College Dataset for Data Modelling 

```python
!bq --location=US mk --dataset college_modeled_vee
```

```python
%%bigquery
# SELECT DISTINCT applies to all columns 
CREATE TABLE college_modeled_vee.Student as
SELECT DISTINCT sid, fname, lname, dob, 'CUR' as status
FROM college_staging_vee.Current_Students
UNION ALL
SELECT DISTINCT sid, fname, lname, CAST(dob as STRING) as dob, 'PRO' as status
from college_staging_vee.New_Students
```

```python
%%bigquery
SELECT * FROM college_modeled_vee.Student 
limit 12
```

```python
%%bigquery
CREATE TABLE college_modeled_vee.Teacher as
SELECT DISTINCT tid, instructor, dept
FROM college_staging.Classes
WHERE tid is not NULL
```

```python
%%bigquery
SELECT * FROM college_modeled_vee.Teacher 
limit 12
```

```python
%%bigquery
CREATE TABLE college_modeled_vee.Class as
SELECT DISTINCT cno, cname, credits
from college_staging.Classes
where cno is not null 
```

```python
%%bigquery
SELECT * FROM college_modeled_vee.Class 
limit 12
```

### Creating Junction Tables for Teachers and Takes 

```python
%%bigquery
CREATE TABLE college_modeled_vee.Teaches as
SELECT DISTINCT tid, cno
from college_staging_vee.Classes
where tid is not null
and cno is not null
```

```python
%%bigquery
select * from college_modeled_vee.Teaches 
limit 12
```

```python
%%bigquery
CREATE TABLE college_modeled_vee.Takes as
SELECT DISTINCT sid, cno, grade
from college_staging_vee.Current_Students
where sid is not null
and cno is not null
```

```python
%%bigquery
select * from college_modeled_vee.Takes
limit 12
```
